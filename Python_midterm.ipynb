```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import json
import os
from IPython.core.display import HTML, display
```

```python
r = requests.post("https://ies-midterm.soulution.rocks/login", json = {"cuni": "88952997"})
```

```python
r.text
r_json = r.json()
```

```python
#My personal code
print(r_json['data']['personal_code'])
```

```python
#storing dataset IDs into single object
dataset_id = r_json['data']['dataset_ids']
dataset_id
```

```python
#here we create an empty dataset
#the we request all the remote datasets through itteration over dataset_id
#finally, we use while to itterate over and over until we get "True" to handle possible error

full_dataset = []
for i in dataset_id:
        r2 = requests.get(f"https://ies-midterm.soulution.rocks/data/{i}")
        r2_json = r2.json()
        while r2_json['message'] == 'Error [503] - Sorry, I just slipped and fell. Thank God, I am made of silicon. You need to work this out.':
            r2 = requests.get(f"https://ies-midterm.soulution.rocks/data/{i}")
            r2_json = r2.json()
        full_dataset.append(r2_json)
            
full_dataset
```

```python
#setting dataframe and inspecting its key
df = pd.DataFrame(full_dataset)
df
df_keys = df.keys()
df_keys
```

```python
#modify dataset into more user-friendly version
df_modified = pd.DataFrame(full_dataset[0]['data']['data'])
df_modified
```

```python
#setting date as an index
df_dateidx = df_modified.melt(id_vars=['Date'], var_name='type', value_name = full_dataset[0]['data']['company']).set_index(['Date','type'])
df_dateidx
```
