```python
import pandas as pd
import numpy as np
import matplotlib as plt
from bs4 import BeautifulSoup
import requests
import json
import os
from IPython.core.display import HTML, display
```

```python
#Scrapping https://scrapethissite.com/pages/simple/
r = requests.get("https://scrapethissite.com/pages/simple/")
BeautifulSoup(r.text, 'lxml')
soup = BeautifulSoup(r.text, 'lxml')
country_names = soup.findAll('h3', {'class':'country-name'})
name = [h3.text.strip() for h3 in country_names]
name
```

```python
#Scrapping https://scrapethissite.com/pages/simple/
r = requests.get("https://scrapethissite.com/pages/simple/")
BeautifulSoup(r.text, 'lxml')
soup = BeautifulSoup(r.text, 'lxml')
span_cap = soup.findAll('span',{'class':'country-capital'})
capital = [span.text.strip() for span in span_cap]
span_pop = soup.findAll('span',{'class':'country-population'})
population = [span.text.strip() for span in span_pop]
span_area = soup.findAll('span',{'class':'country-area'})
area = [span.text.strip() for span in span_area]

countries_table = {}
for i in range(len(name)):
    countries_table[name[i]] = {'country-capital':capital[i], 'country-population':float(population[i]), 'country-area':float(area[i])}
    
df = pd.DataFrame(countries_table).T
df

df_upgrade = df.rename(columns={'country-capital':'Capital City','country-population':'Population','country-area':'Land area'})
df_upgrade
```

```python
#Scrapping IES webpage
website = requests.get("https://ies.fsv.cuni.cz/cs/node/48")
BeautifulSoup(website.text, 'lxml')
soup = BeautifulSoup(website.text, 'lxml')
ac_function = soup.findAll('td',{'class':'peopleTableCellAcf'})
academicfun = [td.text.strip() for td in ac_function]
academicfun
```
